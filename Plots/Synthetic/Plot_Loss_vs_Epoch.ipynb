```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss_array = np.load('./Outputs/loss_model.npy')

print(loss_array.shape)
```

```python
sensing_ratio = 1 + np.arange(80)
n_sensing = len(sensing_ratio)
epochs = 20
n_data = 10
n_pts = epochs * n_data
```

```python
legends = ["Proposed Model"]
```

```python
loss_sensing = np.zeros([n_sensing, epochs])

for n_psi in range(n_sensing):
    loss_psi_range = loss_array[n_psi * n_pts : (n_psi + 1) * n_pts]
    
    fig = plt.figure(figsize = (9, 5))

    
    for n_epoch in range(epochs):
        avg_loss = np.average(loss_psi_range[n_epoch * n_data : (n_epoch + 1) * n_data])
        plt.scatter(n_epoch + 1, avg_loss)
        loss_sensing[n_psi, n_epoch] = avg_loss
        
    plt.title(f"Sensing Ratio - {sensing_ratio[n_psi]}%")
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.grid()
    plt.show()
```

```python
fig = plt.figure(figsize = [8, 6])
plt.plot(np.array(1 + np.arange(epochs), dtype = int), loss_sensing[3, :], marker = "^", label = "Loss")
plt.title("Model Loss vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.savefig("./../Plots/synthetic_loss_vs_epoch.pdf")
plt.show()
```
